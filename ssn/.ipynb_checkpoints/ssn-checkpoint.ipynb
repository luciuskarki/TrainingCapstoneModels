# For each word we then need to pass it into each of the models to determine if its sensetive or not

```python
import re
```

```python
def get_numbers_from(token):
    return "".join(char for char in token if char.isdigit())
```

```python
def is_a_ssn(token):
    only_numbers_token = get_numbers_from(token)
    if len(only_numbers_token)!=9:
        return False
    return True
```

```python
ssn_pattern = (
    r'\b(?!000|666|9\d\d)\d{3}'
    r'-?(?!00)\d{2}'
    r'-?(?!0000)\d{4}\b'
)

phone_pattern = r'\b(?:\(\d{3}\)|\d{3})[-.\s]?\d{3}[-.\s]?\d{4}\b'
```

```python
def is_a_ssn_regex(token):
    return bool(re.search(ssn_pattern, token))
```

```python
def is_a_phone_number(token):
    return bool(re.search(phone_pattern, token))
```

```python
file_path = 'words.txt'
file_contents = get_file_contents(file_path)
tokens = split_on_whitespace(file_contents)
for token in tokens:
    if is_a_ssn_regex(token):
        print(token + " is a ssn")
    if is_a_phone_number(token):
        print(token + " is a phone number")
```
