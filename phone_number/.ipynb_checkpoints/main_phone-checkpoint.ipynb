# Helpers

```python
def get_file_contents(file):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
    return content
```

```python
def split_on_whitespace(pre_split_data):
    return pre_split_data.split()
```

# Specific Code

```python
import re
```

```python
phone_pattern = r'\b(?:\(\d{3}\)|\d{3})[-.\s]?\d{3}[-.\s]?\d{4}\b'
```

```python
def is_a_phone_number(token):
    return bool(re.search(phone_pattern, token))
```

```python
file_path = 'words.txt'
file_contents = get_file_contents(file_path)
tokens = split_on_whitespace(file_contents)
for token in tokens:
    if is_a_phone_number(token):
        print(token + " is a phone number")
```
